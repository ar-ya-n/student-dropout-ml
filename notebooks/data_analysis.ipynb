# The dataset shows class imbalance, where non-dropout cases significantly outnumber dropout cases. This may affect model performance and requires special handling.

```python
df['Dropout'].value_counts().plot(kind='bar')
plt.title("Dropout Distribution")
plt.xlabel("Dropout")
plt.ylabel("Count")
plt.show()
```

```python
for col in df.select_dtypes(include='object').columns:
    print(f"{col}: {df[col].unique()}")
    print()
```

```python
corr = df.corr(numeric_only=True)

plt.figure(figsize=(12, 8))
plt.imshow(corr)
plt.colorbar()
plt.title("Correlation Heatmap")
plt.show()
```

# Correlation analysis revealed moderate relationships between certain features, indicating potential predictors for dropout without severe multicollinearity.
